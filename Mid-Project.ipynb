```python
#check the outliers 
def plot_outliers(numerical_df):
    
    for col in numerical_df.columns:
        plt.figure()
        plt.title(col)
        numerical_df[col].plot(kind='box')

plot_outliers(numerical_df)

#I already replaced the 0, the other utliers are essential for the model so I will leave them 
```

```python
#normalize numerical data
import numpy as np 
from sklearn.preprocessing import Normalizer

X = numerical_df.select_dtypes(include = np.number)
transformer = Normalizer().fit(X)
x_normalized =  transformer.transform(X)
x = pd.DataFrame(x_normalized)
x.columns = X.columns
x.head()
```

```python
#scaling the data, I use StandardScaler to maintain the impact of the outliers
from sklearn.preprocessing import StandardScaler
X_scaled = pd.DataFrame(StandardScaler().fit(x).transform(x))
X_scaled
```

```python
X_scaled.columns = numerical_df.columns
X_scaled.head(1)
```

```python
# Adding constant so the model may act better 
import statsmodels.api as sm
X_scaled_df = sm.add_constant(X_scaled)
X_scaled_df.head()
```

```python
#encoding categorical data - countries
from sklearn.preprocessing import LabelEncoder
data['country_number'] = LabelEncoder().fit(data['country']).transform(data['country'])
encoded_country = data['country_number']
country_df = data['country']
```

```python
encoded_df = pd.concat([X_scaled_df, encoded_country], axis=1)
encoded_df.head()
```

```python
encoded_df.columns
```

```python
#split test 
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X = encoded_df.drop('corruption index',axis=1)
y = encoded_df['corruption index']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
r2_score(y_test, predictions)

#Using this model we may predict 86% of the DataFrame
```

```python
## Mean squared error - the average amount the model is wrong 
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predictions)
print(mse)

# The average difference between Dataset and prediction os 0.14
```

```python
import math
rmse = math.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,predictions)
print (mae)
```
